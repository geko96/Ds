# librerias usadas

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

# Dataset Usado


```python
ds = pd.read_csv("comerce.csv")
df = pd.DataFrame(ds)
df.sample(n=5)
```

```python

# cuenta el número de órdenes por hora del día
orders_by_hour = df.groupby('order_hour_of_day').size()
# crea un gráfico de torta
orders_by_hour.plot(kind='pie', autopct='%1.1f%%')

# personaliza el gráfico
plt.title('Porcentaje de ventas por horario')
plt.ylabel('')
plt.show()
```

```python
# cuenta el número de órdenes por día de la semana
orders_by_day = df.groupby('order_dow').size()

# crea un gráfico de barras
orders_by_day.plot(kind='bar', color='green')

# personaliza el gráfico
plt.title('Ventas según los días de la semana')
plt.xlabel('Día de la semana')
plt.ylabel('Cantidad de órdenes')
plt.xticks(range(7), ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'], rotation=0)
plt.show()
plt.show()
```

## Primer modelo

```python

# eliminar las filas con valores NaN o infinitos
df = df.dropna()

# seleccionar las características para utilizar en el modelo
features = ['order_dow', 'order_hour_of_day', 'days_since_prior_order', 'add_to_cart_order', 'department_id']

# dividir los datos en conjuntos de entrenamiento y prueba
train, test = train_test_split(df, test_size=0.2, random_state=42)

# crear el clasificador de árbol de decisión
clf = DecisionTreeClassifier()

# entrenar el modelo
clf.fit(train[features], train['reordered'])

# predecir los valores para el conjunto de prueba
pred = clf.predict(test[features])

# evaluar la precisión del modelo
accuracy = accuracy_score(test['reordered'], pred)

print(f"La precisión del modelo es {accuracy}")
```
