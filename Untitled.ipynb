```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
```

# Primera entrega

```python
año = "2005"
url = "https://www.basketball-reference.com/leagues/NBA_" + año + "_per_game.html"

df = pd.read_html(url)
dF = df[0].apply(pd.to_numeric, errors='ignore')
dF
```

# Desafio Opcional

```python
dF['Pos'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Cantidad de jugadores por posición')
plt.legend(dF['Pos'].unique())
plt.show()
```

- La mayoria se presentan en posicion PF //  Ala-pívot (Power Forward o PF en EE. UU.)

```python
dF['Age'] = pd.to_numeric(dF['Age'], errors='coerce')
dF['Age'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Cantidad de jugadores por edad')
plt.show()
```

 - Se puede denotar que el grupo de edad mas comun es de 24 años
